Multiplexado Temporal (Temporal Multiplexing): Es una técnica de modulación que permite la transmisión o almacenamiento de múltiples señales en un solo canal o medio, asignando diferentes intervalos de tiempo a cada señal. En el contexto de óptica y holografía, esta técnica implica la grabación de varios hologramas en un mismo medio, cada uno en un momento distinto, de manera que cada holograma puede ser accedido o reproducido en su intervalo de tiempo específico sin interferencia de los otros.
